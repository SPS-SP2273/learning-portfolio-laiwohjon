# What to expect in this chapter

There are many packages (e.g., Matplotlib, Plotty, Seaborn, Bokeh, Folium) that you can use to plot with Python. Of these, Matplotlib is the most popular and most versatile. Since some other packages are built using Matplotlib, it is usually good to know your way around Matplotlib first.

# 1 A simple plot

## 1.1 Letâ€™s look at some code.

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting

#This means that I am plotting y_1 against x with line label
#as "Y values". 
plt.plot(x, y_1, label='Y values',
         color='red', linestyle='dashed')

#Following 2 lines will label the x-axis and y-axis
#respectively. 
plt.xlabel('x-values')
plt.ylabel('y-values')

#Title of the graph 
plt.title('X vs Y')

#alpha refers to the level of transparency of the grid lines
#Alpha takes a value between 0 (completely transparent),
#and 1(ompletely opaque).
plt.grid(alpha=0.5)

#Location of the label (upper right)
plt.legend(loc='upper left')
plt.legend(loc= 'upper right')

#Command given later will take precendent.
```

## 1.3 Yet another plot but with error bars

**Error bars indicate the range within which the true values may lie**

For each data point, an error bar is drawn horizontally with a length of x_error. Horizontal error bars typically represent the uncertainty or variability in the x-coordinate values.

yerr=y_error: This specifies vertical error bars. For each corresponding data point in the y_3 array, an error bar is drawn vertically with a length of y_error. Vertical error bars typically represent the uncertainty or variability in the y-coordinate values.

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

#This will plot y_3 against x with error bars incorporated. 
plt.errorbar(x, y_3,
             xerr=x_error, yerr=y_error,
             label="Y3 with errors",
             color="green")
```

# 2 Better with NumPy

Often, it is easier to use NumPy arrays instead of Python lists. So, letâ€™s first convert the Python lists into NumPy arrays and then redo the plot in the previous step.

```python
import numpy as np

# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
```

You must wait until the next step to see why NumPy is better.

## 2.1 Adding mathematical functions

One of the advantages of NumPy arrays is that they allow us to generate data-related mathematical functions easily **(recall that we can change each element according to some function more easily using NumPy arrays)**. Letâ€™s reuse our previous code to plot $ x^2 $ and $sin(x)$

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  # The math stuff
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()                                 
```

Alas, our plot does not look good because 
sin(x) lies between -1 and 1, but $x^2$ has no such bounds. One way to fix this is to add another y-axis that shares the same x-axis.



## We need another axis!

Matplotlib offers a variety of ways to have multiple axes. The simplest way is to have another y-axis that shares the same x-axis. We can use the command `twinx()` for this.

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

#The new y axis will lead the y_values of the new plot 
#of sin_x to span the whole vertical axis
#instead of being bunched up near 0. 
plt.twinx()                                                   # This creates a new y-axis 

                                                            # for the plots that comes after
# We plot new graph against the new axis
#location of legend alpplies to the plot directly before it
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
```

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
#Next line generates 1000 points between the min and max x-val
#This ensures that curve drawn will be smoother as the
#number of points inceases. 
x_smooth = np.linspace(x.min(), x.max(), 1000)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
sin_x_smooth = np.sin(x_smooth)
plt.plot(x_smooth, sin_x_smooth, label='sin(x)',color='blue', )
plt.legend(loc='lower right')         # For y-axis 2
```

### Things to note

- We now have two legend() calls, one for each axis.
- our plot still does not look good without np.linspace because we have only a few points. We can use np.linspace(x.min(), x.max(), *large number*) to fix

# 3 Saving to disc

If you want to use your plot in a report or presentation, you must first save it to disk. Luckily, Matplotlib makes it astonishingly easy to export plots into many formats (PDF, JPEG, PNG, BMP, etc.) and at different resolutions.

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')

plt.savefig('simple-plot.png', dpi=150)
```

**savefig()**: Does the thing its name implies

**simple-plot.png**: filename.extension (we need to specify extension)

**dpi**: dots per inch (used to specify resolution) - Americans and their imperial units haiz

When you run the above code, you will find the file saved in the same directory (folder) as the one in your notebook lives in. If you want it saved elsewhere, specify the path in more detail. 

Example: `plt.savefig('C://Desktop/simple-plot.png', dpi=150)`

# 4 A real example: Global Warming

## 4.1 Plotting data from files

use the Earthâ€™s land temperature data from the Berkeley Earth website. Please visit the site (Global Warming 
 Data Overview) and download the average temperature data for Daily Land. The original name of the file should be Complete_TAVG_daily.txt

```python
#We skip first 24 rows using skiprows as data only starts
#at the 25th row. 
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)

#Scan through all rows but use only the 0th column.
date = data[:, 0]

#Anomaly data is found in the rightmost column
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])

# We add ahorizontal line at the zero value to highlight
# the trend shown by the data.
#'hlines' stands for horizontal line.

#date[0] and date[-1] are the starting and ending 
#x-values of the line. 
plt.hlines(0, date[0], date[-1], linestyle='-', colors='grey')
```

## 4.2 Styles

We might want a decent-looking plot without spending time customising it. To facilitate this Matplotlib offers some standard style templates. I am going to use the one called fivethirtyeight.

```python
plt.style.use('fivethirtyeight')

#We skip first 24 rows using skiprows as data only starts
#at the 25th row. 
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)

#Scan through all rows but use only the 0th column.
date = data[:, 0]

#Anomaly data is found in the rightmost column
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])

# We add ahorizontal line at the zero value to highlight
# the trend shown by the data.
#'hlines' stands for horizontal line.

#date[0] and date[-1] are the starting and ending 
#x-values of the line. 
plt.hlines(0, date[0], date[-1], linestyle='-', colors='grey')
```

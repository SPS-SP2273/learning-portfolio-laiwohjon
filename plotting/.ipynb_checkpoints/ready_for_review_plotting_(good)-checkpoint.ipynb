# What to expect in this chapter

So far, the ‘dialect’ we have used to talk to Matplotlib is called the Matlab-like **pyplot(plt) interface**. From here onward, I will show you how to use the other, more powerful ‘dialect’ called the **Object Oriented (OO)** interface. This way of talking to Matplotlib gives us a more nuanced control over what is going on by allowing us to manipulate the various axes easily.

# 1 Some nomenclature

Before going ahead, let’s distinguish between a Matplotlib figure and an axis.

A figure is simple; it is the full canvas you use to draw stuff on. An axis is the individual mathematical axes we use for plotting. So, one figure can have multiple axes, as shown below, where we have a (single) figure with four axes.


![](https://sps.nus.edu.sg/sp2273/docs/python_basics/07_plotting/__imgs__/plotting_good_exercise-01_solution.png)

By the way, we had already encountered a situation with multiple axes in the last chapter when we used twinx()

# 2 Comparing the two ‘dialects’

Lets create the same plot using both pyplot and OO ‘dialects’ so that we can see how the latter works. 

![](https://sps.nus.edu.sg/sp2273/docs/python_basics/07_plotting/__imgs__/plotting_good-01.png)

As we can see, both sets of code below will produce the same plot. 

### pyplot Interface

```python
import numpy as np
from matplotlib import pyplot as plt
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)

plt.plot(x, cos_x, label='cos x')
plt.plot(x, sin_x, label='sin x')
plt.legend()

```

### OO Interface

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x, cos_x, label='cos x')
ax.plot(x, sin_x, label='sin x')
ax.legend()
```

For the OO interface, we have to start by using subplots() to ask Matplotlib to create a figure and an axis. Matplotlib obliges and gives us a figure (`fig`) and an axis (`ax`).

Note: We don't actually need to call these variables fig and ax, but it is commonly used in documentation. In this example, I need only one column and one row, but we can have more. 

Yes, the OO looks more comlicated than the pyplot version. But, it offers so much freedom that it is worth learning it for more demanding, complex plots. 

# 3 What is OO ax, really?

The code below creates the (crude) plot. 

```python
fig, ax = plt.subplots(nrows=3, ncols=1)

ax[0].plot(x, cos_x, label='cos x')
ax[1].plot(x, sin_x, label='sin x')

ax[0].legend()
ax[1].legend()
```

To get the above plot, we must ask for three rows (`nrows=3`) and one column (`ncols=1`). We do this by using subplots()

`fig, ax = plt.subplots(ncols=1, nrows=3)`

This should give me two axes so that I can plot in both panes. Let’s quickly check a few more details about ax.

```python
type(ax)
```

We can see that ax is a NumPy array.

```python
ax.shape
```

NumPy has 3 "things", corresponding to 3 axes. 

```python
type(ax[0])
```

ax[0] is just a Matplotlib axis.

# 4 A complete OO example

The following is a simple example that creates a nicer, tweaked version of the previous plot.


```python
# We Set up the axes and figure, get two axes ready to plot in two rows in one column.
#Change the size of the figure by specifying a figure size
#Ask that the plots share the x-axis using sharex.
fig, ax = plt.subplots(nrows=2, ncols=1,
                       figsize=(5, 5),
                       sharex=True)

#Since ax is a NumPy array with two axes, we can index them using 0 and 1. Then we just plot to that axis.
# Use fill_between() to fill the plots. Use the indexed 'ax' to access the axes.
ax[0].plot(x, cos_x, label=r'$\cos(x)$')
ax[0].fill_between(x, 0, cos_x, alpha=.25)
ax[1].plot(x, sin_x, label=r'$\sin(x)$')
ax[1].fill_between(x, 0, sin_x, alpha=.25)

#Draw legend (can use for loop to save time)
#Alpha is the opacity of grid.
for a in ax:
    a.legend()
    a.grid(alpha=.25)
    a.set_ylabel('$y$', rotation=0)

#We are sharing the x-axis. So, we only need to label the lowest plot. This has the index 1.
ax[1].set_xlabel('$x$')

#let’s add a super title to the figure (notplot)
#ask Matplotlib to make any necessary adjustments to the layout to make our plot look nice by calling tight_layout().
fig.suptitle(r'$\sin(x)$ and $\cos(x)$')
fig.tight_layout()
```

Unfortunately, the pyplot and OO interfaces don’t use identical function names. For example, pyplot used xlabel() to set the x label, but OO uses set_xlabel().

# 5 Multiple rows and columns

### Using ax

```python
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(5, 5),
                       sharex='col', sharey='col')

# Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
```

## Using ax

First part of the above code uses ax. We shall now discuss how to use ax.

```python
ax.shape
```

ax, as defined in the code above, is a 2x2 array. 

We can set the tittle of each of the plots separately as follows:

```python
ax[0, 0].set_title('Top Left')
ax[0, 1].set_title('Top Right')
ax[1, 0].set_title('Bottom Left')
ax[1, 1].set_title('Bottom Right')
```

We can also assign each element in array to familiar variables:

```python
top_left=ax[0, 0]
top_right=ax[0, 1]
bottom_left=ax[1, 0]
bottom_right=ax[1, 1]
```

You can also use:

```python
top_left, top_right, bottom_left, bottom_right = ax.flatten()
```

flatten() takes the 2D array and ‘flattens’ it into a 1D array; unpacking takes care of the assignments (recall earlier chapters).

To draw the lines, we use the very useful Matplotlib function hlines():

```python
top_left.hlines(y=1, xmin=0, xmax=5)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=5)
bottom_right.hlines(y=4, xmin=0, xmax=5)
```

The xmin and xmax just tells us where the horizontal line starts and where it ends. 

## 5.1 Sharing axes

We asked Matplotlib to make the plots more compact by sharing the x and y axes using sharex and sharey.

Lets see what happens when we do not specify how to share. 

```python
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(5, 5))

# Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
```

Matplotlib has auto-scaled both axes. In particular, the plots on the left go from 0 to 4, and those on the right go from 0 to 5, as these are the lengths we used for the lines. 

Also now, the lines all have the same vertical distance above and below the lines. 

We can specify how to share the axes in 3 ways:
- True: Make all the axes use the same range
- col: Use the same range for all the columns.
- row: USe the same range for all the rows.

```python
fig, ax = plt.subplots(
    nrows=2, ncols=2,
    figsize=(5, 5),
    sharex=True, sharey='row'
)

# Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
```

For the above graph, all plots have the same range for the x-axis and plots in the same row share the same y-axis.
